## First Thoughts

We may immediately think of using a for loop to iterate over all $100$ integers, but this presents a few issues: 
- it is computationally intensive, though storing these values can help;
- it requires us to just crunch numbers, and with certain languages could lead to errors for suitably high values;
- we learn nothing new

```python
targetNum = 100
```

```python
factorialStorage = [1]*(targetNum+1)
for i in range(2, targetNum+1):
    factorialStorage[i] = (i)*factorialStorage[i-1]
facTarget = factorialStorage[targetNum]
print(facTarget)

facStr = str(facTarget)
facNum = [int(i) for i in facStr]
print(facNum)  

count = 0
for char in facNum:
    count += facNum[char]
print(count)
```

## Improved Solution

Rather than compute the final integer using just a list, we can instead try to reconstruct it using factors!

To do so, we must find all primes below the target number. Knowing that, we also know that the prime $p$ shows up exactly:

$$\#(p) =\left \lceil{\frac{n}{p}}\right \rceil + \left \lceil{\frac{n}{p^2}}}\right \rceil + \dots$$
