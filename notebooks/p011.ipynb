Easy problem, cuts work in half when you realise scanning leftwards and rightwards is redundant, alongside (NE, SW), (NW, SE), (Up, Down). Then can define a right-scanning and diagonal-scanning method, and by transposing or flipping the number grid we can just reuse these methods. Initially 8 directions, reduced down to 2. 

```python
numGrid = [
    [int(x) for x in row.split()]
    for row in rawGrid.strip().splitlines()
]
```

```python
def rightScan(grid: list[list[int]] = numGrid, depth:int = 4) -> int:
    rightProduct = 0

    for row in grid:
        for colIndex in range(len(row)- depth +1):
            product =  1

            for x in range(depth):
                product *= row[colIndex+ x]

            rightProduct = max(rightProduct, product)
    return rightProduct

rightMax = rightScan()
print(rightMax)
```

```python
def transpose(grid:list[list[int]] = numGrid)-> list[list[int]]:
    return [list(row) for row in zip(*grid)]
```

```python
downMax = rightScan(transpose(numGrid))
print(downMax)
```

```python
def diagonalScan(grid: list[list[int]] = numGrid, depth:int = 4) -> int:
    diagProduct = 0
    for rowIndex in range(len(grid[0])-depth+1):
        for columnIndex in range(len(grid[0])-depth+1): #works since square, assume that all equal, will add check earlier
            product = 1

            for x in range(depth):
                product *= grid[rowIndex+x][columnIndex+x]
            
            diagProduct = max(diagProduct, product)
    return diagProduct

diagMax = diagonalScan()
print(diagMax)
```

```python
def flipLeftRight(grid: list[list[int]] = numGrid) -> list[list[int]]:
    return [row[::-1] for row in grid]
```

```python
reverseDiagMax = diagonalScan(flipLeftRight(numGrid))
print(reverseDiagMax)
```

```python
TotalMax = max(diagMax, reverseDiagMax, rightMax, downMax)
print(TotalMax)
```
