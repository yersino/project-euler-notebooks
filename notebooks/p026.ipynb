But wait, this doesn't work! Python doesn't store these decimals to arbitrary precision, and we have no guarantee that the entire repeating cycle has been stored! 

Instead we have to try another approach. 

Instead, we can simulate the long division algorithm by hand. The key observation is that the decimal expansion begins to repeat precisely when a remainder repeats.

```python
def cycleFinder(num: int) -> tuple[int, list[int], int]:
    digits = []
    seenRemainders = {}

    remainder = 1 % num
    position = 0
    cycleLength = 0

    while remainder not in seenRemainders and remainder != 0:
        seenRemainders[remainder] = position

        digit = (remainder * 10) // num
        digits.append(digit)

        remainder = (remainder * 10) % num
        position += 1

    if remainder == 0:
        cycleStart= position
        cycleLength = 0
    else:
        cycleStart = seenRemainders[remainder]
        cycleLength = position - cycleStart
    return cycleLength, digits, cycleStart
```

```python
longestCycle = 0
bestNum = 0

for num in range(2, numTarget):
    length, digits, cycleStart = cycleFinder(num)

    if length > longestCycle:
        longestCycle = length 
        bestNum = num
        bestDigits = digits
        bestCycleStart = cycleStart

nonRepeating = ''.join(map(str, bestDigits[:bestCycleStart]))
repeating = ''.join(map(str, bestDigits[bestCycleStart:]))

print(f"1/{bestNum} = 0.{nonRepeating}({repeating})")
```

And we get an answer of $983$.
