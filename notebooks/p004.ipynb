## Brute force approach

Have an upper bound of $999*999 = 998001$. So anticipate that the largest palindrome made from product of two $3-$ digit numbers will have $6$ digits. 

First idea: 
Iterate over all initial $3-$ digit start codons, checking factorisations over $999, 998, 997$. But this requires some sort of factorisation code to check if there are factors within the range $[100,999]$.

Second idea:
Brute force all product pairs for factors $f, F$ in the range $[100, 999]$, store the running total and just check if each product is a palindrome. Requires simpler code to just check if a number is a palindrome. This would give at most $900*900 = 810000$ combinations, which is reasonable for brute-force.

First optimisation is to say WLOG that $f \leq F$ to avoid reduce the number of pairs we have to check by a factor of $2$.

We get that the maximal palindromic product is $906609 = 913*993$. Our code ran in $<0.1$ seconds.

```python
def checkPalindrome(nums: int) -> bool:
    s = str(nums)
    return s == s[::-1]
```

```python
def bruteForceSolution()->int:
    highest = 0
    highestFactors = None
    for f in range(100, 1000):
        for F in range(f, 1000):
            product = f*F
            if product > highest and checkPalindrome(product):
                highest = product
                highestFactors = (f, F)
    return highest, highestFactors

bruteForceSolution()
```


## Complexity

Let $n$ be the number of candidate factors. Here, $n = 900$.

The brute force method checks roughly

$$
\frac{n^2}{2}
$$

products, so the time complexity is

$$
O(n^2).
$$

The space complexity is

$$
O(1),
$$

since we only store the current best product and its factors.



## Mathematical observation
A six-digit palindrome has the form

$$
abccba.
$$

Writing this as a number,

$$
abccba = 100000a + 10000b + 1000c + 100c + 10b + a.
$$

Collecting terms gives

$$
abccba = 100001a + 10010b + 1100c.
$$

Factorising,

$$
abccba = 11(9091a + 910b + 100c).
$$

So every six-digit palindrome is divisible by $11$.

This means that if the largest palindrome product is six digits, then at least one of its two three-digit factors must be divisible by $11$.

## Lessons learned
- A simple brute force solution is often good enough for small search spaces.
- Symmetry can reduce repeated work.
- Number-theoretic structure can improve a brute force search.
- Six-digit palindromes are always divisible by $11$.
